The correlation matrix is used to deteremine the releationships between features to decide the best approach to deal with missing values.

```python
corr_matrix = vgsales_df.corr()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f',linewidth=.10)
plt.title('Correlation Matrix')
plt.show()
```

To have a better understanding of the distribution of null values, the isnull visualisation is used to understand the distribution of missing values and make a decision on whether to drop columns or rows that containn a high number of missing values or use another technique to deal with missing values.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(vgsales_df.isnull(), cmap='viridis', cbar=False)
plt.title('Distribution of Missing Values')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

The following columns have been dropped as they contain a large number of missing values and they are irrelevant to the required analysis and the correlation matrix shows lower releationship with the target variable. Also dropping the irrelevant columns helps reduce the complexity as the fill in missing values can produce bias. User Count is kept to be used in part two of the analysis.

```python
vgsales_df.drop(['Critic_Score','Critic_Count','User_Score','Rating'],axis=1,inplace=True)
```

```python
vgsales_df.dropna(subset=['Name','Genre'],axis=0,inplace=True)
```

```python
from sklearn.impute import KNNImputer
```

```python
columns_to_impute = ["Year_of_Release", "User_Count"]
```

```python
knn_imputer = KNNImputer()
```

```python
cleansales_df = knn_imputer.fit_transform(vgsales_df[columns_to_impute])
```

```python
vgsales_df[columns_to_impute] = cleansales_df
```

```python
vgsales_df[columns_to_impute]
```

```python
cleansales_df = pd.DataFrame(cleansales_df, columns=columns_to_impute)
cleansales_df
```

```python
vgsales_df.dropna(subset=['Publisher'],axis=0,inplace=True)
```

```python
vgsales_df.isnull().sum()
```

```python
vgsales_df.shape
```

```python
vgsales_df['Critic_Score'].fillna(vgsales_df['Critic_Score'].mean(), inplace=True)
```

```python
vgsales_df['Critic_Count'].fillna(vgsales_df['Critic_Count'].median(), inplace=True)
```

```python
vgsales_df['User_Score'].unique()
```

```python
vgsales_df['User_Score'].fillna(vgsales_df['User_Score'].mode()[0],inplace=True)
```

```python
vgsales_df['User_Count'].fillna(vgsales_df['User_Count'].median(),inplace=True)
```

```python
vgsales_df['Developer'].fillna(method='ffill',inplace=True)
```

```python
vgsales_df['Rating'].fillna(method='ffill',inplace=True)
```

```python
vgsales_df.dropna(subset='Name',axis=0,inplace=True)
```

```python
sales_2010_2014 = vgsales_df[(vgsales_df['Year_of_Release'] >= 2010) & (vgsales_df['Year_of_Release'] <= 2014)]
```

```python
correlation = sales_2010_2014['NA_Sales'].corr(sales_2010_2014['JP_Sales'])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(sales_2010_2014['NA_Sales'], sales_2010_2014['JP_Sales'], alpha=0.5)
plt.title('Correlation between NA Sales and JP Sales (2010-2014)')
plt.xlabel('NA Sales')
plt.ylabel('JP Sales')
plt.text(10, 6, f'Correlation coefficient: {correlation:.2f}', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
plt.grid(True)
plt.show()
```

```python
older_games = vgsales_df[vgsales_df['Year_of_Release'] <= 2005]
newer_games = vgsales_df[vgsales_df['Year_of_Release'] > 2005]
```

```python
mean_eu_sales_old = older_games['EU_Sales'].mean()
mean_eu_sales_new = newer_games['EU_Sales'].mean()
```

```python
print("Mean EU_Sales for older games (2005 and earlier):", mean_eu_sales_old)
print("Mean EU_Sales for newer games (after 2005):", mean_eu_sales_new)
```

```python
if mean_eu_sales_old > mean_eu_sales_new:
    print("Older games have a higher mean EU_Sales than newer games.")
elif mean_eu_sales_old < mean_eu_sales_new:
    print("Newer games have a higher mean EU_Sales than older games.")
else:
    print("The mean EU_Sales for older and newer games are the same.")
```

```python
plt.figure(figsize=(8, 6))
plt.bar(['Older Games (<= 2005)', 'Newer Games (> 2005)'], [mean_eu_sales_older, mean_eu_sales_newer], color=['skyblue', 'lightgreen'])
plt.title('Mean EU Sales: Older vs. Newer Games')
plt.ylabel('Mean EU Sales')
plt.ylim(0, max(mean_eu_sales_older, mean_eu_sales_newer) * 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
